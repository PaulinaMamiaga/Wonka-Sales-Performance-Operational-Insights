````python
# Initialize dataframe totals
totals = pd.DataFrame()

# Calculate total sales by product name
totals["sales"] = df.groupby('Product Name')['Sales'].sum()
print("\nTotal Sales by Product Name -------------------------")
print(totals["sales"])

# Calculate total cost by product name
totals["cost"] = df.groupby('Product Name')['Cost'].sum()
print("\nTotal Cost by Product Name -------------------------")
print(totals["cost"])

# Calculate total gross profit by product name
totals["gross_profit"] = df.groupby('Product Name')['Gross Profit'].sum()
print("\nTotal Gross Profit by Product Name -------------------------")
print(totals["gross_profit"])

# Calculate total units by product name
totals["units"] = df.groupby('Product Name')['Units'].sum()
print("\nTotal Units by Product Name -------------------------")
print(totals["units"])

# Make "Product Name" a column instead of the index
totals = totals.reset_index()  # <-- this is the key step

# Optional: sort by gross profit descending
totals = totals.sort_values(by="gross_profit", ascending=False)

# Save to CSV Product Totals
totals.to_csv("../data/interim/product_totals.csv", index=False)

# View the dataframe
totals.head(totals.shape[0])
````

````python
import pandas as pd

# total gross profit per customer
gross_profit_by_customer = (
    df.groupby("Customer ID", as_index=False)["Gross Profit"]
      .sum()
      .rename(columns={"Gross Profit": "Total Gross Profit"})
)

# sort descending by total gross profit
gross_profit_by_customer = gross_profit_by_customer.sort_values(
    by="Total Gross Profit",
    ascending=False
)

# print full table
pd.set_option("display.max_rows", None)   # show all rows
print("Customer ID - Total Gross Profit table:")
print(gross_profit_by_customer.head())

# best customer
best_customer = gross_profit_by_customer.iloc[0]

# worst customer
worst_customer = gross_profit_by_customer.iloc[-1]

print("\nBest customer ID:", int(best_customer["Customer ID"]),
      "Total Gross Profit:", best_customer["Total Gross Profit"])

print("Worst customer ID:", int(worst_customer["Customer ID"]),
      "Total Gross Profit:", worst_customer["Total Gross Profit"])
````

````python
import pandas as pd

# total gross profit per customer
gross_profit_by_customer = (
    df.groupby("Customer ID", as_index=False)["Gross Profit"]
      .sum()
      .rename(columns={"Gross Profit": "Total Gross Profit"})
      .sort_values("Total Gross Profit", ascending=False)
)

best_customer  = gross_profit_by_customer.iloc[0]["Customer ID"]
worst_customer = gross_profit_by_customer.iloc[-1]["Customer ID"]

print("Best customer ID:", int(best_customer))
print("Worst customer ID:", int(worst_customer))

# products most purchased (by units) for best customer
best_products = (df[df["Customer ID"] == best_customer]
                 .groupby("Product Name")["Units"]
                 .sum()
                 .sort_values(ascending=False))

# products most purchased (by units) for worst customer
worst_products = (df[df["Customer ID"] == worst_customer]
                  .groupby("Product Name")["Units"]
                  .sum()
                  .sort_values(ascending=False))

print("\nBest customer's products (by units):")
print(best_products)

print("\nWorst customer's products (by units):")
print(worst_products)
````
