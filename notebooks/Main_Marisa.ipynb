````python
# Initialize dataframe totals
totals = pd.DataFrame()

# Calculate total sales by product name
totals["sales"] = df.groupby('Product Name')['Sales'].sum()
print("\nTotal Sales by Product Name -------------------------")
print(totals["sales"])

# Calculate total cost by product name
totals["cost"] = df.groupby('Product Name')['Cost'].sum()
print("\nTotal Cost by Product Name -------------------------")
print(totals["cost"])

# Calculate total gross profit by product name
totals["gross_profit"] = df.groupby('Product Name')['Gross Profit'].sum()
print("\nTotal Gross Profit by Product Name -------------------------")
print(totals["gross_profit"])

# Calculate total units by product name
totals["units"] = df.groupby('Product Name')['Units'].sum()
print("\nTotal Units by Product Name -------------------------")
print(totals["units"])

# Make "Product Name" a column instead of the index
totals = totals.reset_index()  # <-- this is the key step

# Optional: sort by gross profit descending
totals = totals.sort_values(by="gross_profit", ascending=False)

# Save to CSV Product Totals
totals.to_csv("../data/interim/product_totals.csv", index=False)

# View the dataframe
totals.head(totals.shape[0])
````

````python
# Profitability metrics per product
totals["profit_margin"] = totals["gross_profit"] / totals["sales"]
totals["profit_per_unit"] = totals["gross_profit"] / totals["units"]

# Keep a nicely ordered table for inspection
product_profit = totals.sort_values("profit_margin", ascending=False)
product_profit
````

````python
# Best product by profit margin
best = product_profit.iloc[0]
print("BEST PRODUCT (by profit margin)")
print(best[["Product Name", "sales", "gross_profit", "profit_margin", "profit_per_unit"]])

# Worst product by profit margin
worst = product_profit.iloc[-1]
print("\nWORST PRODUCT (by profit margin)")
print(worst[["Product Name", "sales", "gross_profit", "profit_margin", "profit_per_unit"]])
````

````python
# Best and worst by gross profit
product_profit_gp = totals.sort_values("gross_profit", ascending=False)
best_gp = product_profit_gp.iloc[0]
worst_gp = product_profit_gp.iloc[-1]
print("BEST PRODUCT (by gross profit)")
print(best_gp[["Product Name", "sales", "gross_profit", "profit_margin", "profit_per_unit"]])
````

````python
# Plot total gross profit by product
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(product_profit["Product Name"], product_profit["gross_profit"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("Total Gross Profit")
plt.title("Total Gross Profit by Product")
plt.tight_layout()
plt.show()
````

````python
# Plot profit margin vs sales
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# Sort by profit margin (best to worst)
pp_margin = product_profit.sort_values("profit_margin", ascending=True)

plt.figure(figsize=(8, 6))
plt.barh(pp_margin["Product Name"], pp_margin["profit_margin"])

ax = plt.gca()
ax.xaxis.set_major_formatter(PercentFormatter(1.0))  # 0–1 -> 0–100%

plt.xlabel("Profit Margin (%)")
plt.title("Profit Margin by Product")
plt.tight_layout()
plt.show()
````

````python
# Compute each product's share of total gross profit
pp_gp = product_profit.copy()
pp_gp["profit_share"] = pp_gp["gross_profit"] / pp_gp["gross_profit"].sum()
pp_gp = pp_gp.sort_values("profit_share", ascending=True)

plt.figure(figsize=(8, 6))
plt.barh(pp_gp["Product Name"], pp_gp["profit_share"])

ax = plt.gca()
ax.xaxis.set_major_formatter(PercentFormatter(1.0))  

plt.xlabel("Share of Total Gross Profit (%)")
plt.title("Share of Total Gross Profit by Product")
plt.tight_layout()
plt.show()
````

## Summary of findings for: "Which product performs the best vs the worst in termas of profitabiliy?"

### Best products with high margin + high total profit contribution
These products dominate both profit margin and share of total gross profit:
- Wonka Bar – Scrumdiddlyumptious
- Wonka Bar – Triple Dazzle Caramel
- Wonka Bar – Milk Chocolate
- Wonka Bar – Nutty Crunch Surprise
- Wonka Bar – Fudge Mallows

These 5 products each contribute ~17–21% of total gross profit, making them the core profit drivers.
Their margins are also strong, consistently around 65–70%, which reinforces their role as “premium profitability products.”

#### Recommentation:
Continue investing and expanding production for these top sellers. Their high margin + high total profit combo makes them ideal to scale.



### Medium performers with good margin but low contribution
Products like:
- Laffy Taffy
- Fizzly Lifting Drinks
- Wonka Gum
- Lickable Wallpaper

show healthy profit margins (50–60%), but contribute very little to total profit due to low sales volume.

#### Recommentation:
Increase marketing and distribution! These products are profitable per unit but underperform in volume. Improving visibility could boost their total contribution.



### Worst products with low margin and low total profit
The weakest performers are:
- Kazookles → lowest margin (~5–10%) and almost zero total gross profit
- Fun Dip → ~40% margin but extremely low total profit
- Nerds, SweeTARTS, Everlasting Gobstopper, Hair Toffee → strong margin but negligible contribution

Low-margin, low-sales products may be priced incorrectly, poorly positioned, or not meeting consumer demand.

#### Recommentation:
- Evaluate pricing strategy for Fun Dip and Kazookles.
- Consider discontinuation or reformulation for products consistently underperforming.
- Reallocate resources toward higher-impact products.
